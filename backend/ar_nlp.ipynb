```python
stemmer = ISRIStemmer()
# Stem the word "تعلمت" (meaning "I learned")
stemmed_word = stemmer.stem("تعلمت")
print(stemmed_word)
# Output: علم
# Stem the word "تعلموا" (meaning "you (plural) learned")
stemmed_word = stemmer.stem("اعلاميون")
print(stemmed_word)
# Output: علم
stemmed_word = stemmer.stem("حركات")
print(stemmed_word)
```
